```python
import pandas as pd
import folium

# CSV 파일에서 데이터 읽어오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/log_parsed.csv')
```

```python
# 지도 초기 설정
m = folium.Map(location=[36.10311, 129.38795], zoom_start=17)

# 마커 추가
def add_marker(row):
    # 정보 팝업 설정
    if pd.isna(row['Signal_Level(-100)']) or pd.isna(row['Link_Quality(/70)']):
        info = "네트워크 연결 끊김"
    else:
        info = f"<p><strong>Signal Level:</strong> {row['Signal_Level(-100)']} dBm</p> \
                <p><strong>Link Quality:</strong> {row['Link_Quality(/70)']}</p>"

    html = f"<div style='padding: 0px; width: 120px;'> \
                <p><strong>Time:</strong> {row['Time']}</p> \
                {info} \
             </div>"

    # 마커 색상 설정
    color = 'blue'
    if pd.isna(row['Signal_Level(-100)']) or pd.isna(row['Link_Quality(/70)']):
        color = 'gray'  # 연결 끊김 상태일 때는 회색으로 표시
    elif int(row['Link_Quality(/70)']) < 35 and int(row['Signal_Level(-100)']) < -70:
        color = 'red'

    # CircleMarker 추가
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=1,  # 점의 반지름 설정
        color=color,  # 점의 색상 설정
        fill=True,
        fill_color=color,  # 점을 채우는 색상 설정
        fill_opacity=0.5,  # 채우는 색상의 투명도 설정
        popup=folium.Popup(html, max_width=300)  # 팝업 설정
    ).add_to(m)

# DataFrame 순회하며 마커 추가
for index, row in df.iterrows():
    add_marker(row)

m
```

```python
# 지도를 HTML 파일로 저장하거나 출력
m.save('/content/drive/MyDrive/Colab Notebooks/map.html')
```
