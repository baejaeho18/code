```python
import numpy as np

a = np.array([1,2,3,4,5])
b = np.array(['대한민국', '포르트갈', '가나', '우루과이'])
c = np.array([1,2,'대한민국', '포르투갈'])

print(a)
print(b)
print(c)  # 숫자와 문자열을 혼합할 경우, 모두 문자열로 처리
```

```python
import pandas as pd

s = pd.Series(['대한민국', '포르투갈', '가나', '우루과이'],
              index=['가', '나', '다', '라'],
              name="2022 카타르월드컵 H조")
print(s)

# Dictionary 형으로 선언: Index - Column
a1 = pd.DataFrame({"a":[1,2,3], "b":[4,5,6], "c":[7,8,9]})
print(a1)

# List 형으로 선언: Row - Index
a2 = pd.DataFrame([[1,2,3], [4,5,6], [7,8,9]],
                  ["a", "b", "c"])
print(a2)
```
